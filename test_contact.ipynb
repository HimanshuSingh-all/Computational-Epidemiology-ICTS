```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
system_size = 1000
arr = np.zeros(system_size)
indices = [i for i in range(system_size)]
```

```python

#define the fraction of the infected in the population
infratio = 0.3
#Populate the exact fraction of the array as 1 meaning infected
indsi = np.random.choice(indices, size= int(system_size*infratio), replace = False)
arr[indsi] = 1

# r = 0.1
# indsi = np.random.choice([i for i in range(system_size)], size= int(system_size*i), replace = False)
# arr[indsi] = 1
# arr
```

```python
contacts = [i for i in range(1, 12)]
probability_infection = list()
infection_prob = 0.2
trials = 10000
for cont in contacts:
    states = list()
    for i in range(trials):
        state = 0
        choseninds = np.random.choice(indices, size = cont, replace = False)
        for ind in choseninds:
            if arr[ind] == 1:
                randomnum = np.random.uniform()
                #a person gets infected with a certain probability
                #afeter coming into contact with an infected person
                if randomnum<infection_prob:
                    state =1
                    break
        states.append(state)
    probability_infection.append(sum(states)/trials)

probability_predicted = 1- (1-infection_prob)**(infratio*np.array(contacts))
```

```python
plt.scatter(contacts, probability_infection, label = 'Simulation')
plt.scatter(contacts, probability_predicted, marker = 'x', label = 'Predicted ')
plt.ylabel('Probability of contracting an infection')
plt.xlabel('Number of contacts per day')
plt.legend()
```
